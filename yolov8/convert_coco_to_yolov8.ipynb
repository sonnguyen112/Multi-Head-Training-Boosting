```python
import json
from PIL import Image
import uuid
```

```python
data = json.load(open("/home/sonnguyen112/Data/HCMUS/Thesis/ByteTrack/datasets/mix_det/annotations/train.json", "r"))
data
```

```python
image_root = "/home/sonnguyen112/Data/HCMUS/Thesis/ByteTrack/datasets/mix_det/"
save_folder = "dataset/mix_det/train/images"
label_forder = "dataset/mix_det/train/labels"
i = 0
for image in data["images"]:
    print("image name: ", image["file_name"])
    img = Image.open(image_root + image["file_name"])
    save_file_name = str(uuid.uuid4()) + "." + image["file_name"].split(".")[-1]
    img.save(save_folder + "/" + save_file_name)
    bboxs = [i for i in data["annotations"] if i["image_id"] == image["id"]]

    with open(label_forder + "/" + save_file_name.split(".")[0] + ".txt", "w") as f:
        for bbox in bboxs:
            cls = bbox["category_id"] - 1
            x_min, y_min, w, h = bbox["bbox"]
            x_center = x_min + w / 2
            y_center = y_min + h / 2
            x_center /= img.width
            y_center /= img.height
            w /= img.width
            h /= img.height
            x_center = min(max(0, x_center), 1)
            y_center = min(max(0, y_center), 1)
            w = min(max(0, w), 1)
            h = min(max(0, h), 1)
            f.write(f"{cls} {x_center} {y_center} {w} {h}\n")
```
