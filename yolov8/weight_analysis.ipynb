```python
import torch
```

```python
modelA = torch.load('yolov8s.pt', map_location=torch.device('cpu'))
```

```python
state_dict_A = modelA["model"].state_dict()
```

```python
state_dict_A.keys()
```

```python
modelB = torch.load('yolov8x.pt', map_location=torch.device('cpu'))
```

```python
state_dict_B = modelB["model"].state_dict()
```

```python
state_dict_B.keys()
```

```python
for k in list(state_dict_B.keys()):
    if "22" in k:
        state_dict_A[k.replace("22", "26")] = state_dict_B[k]
        state_dict_A[k.replace("22", "33")] = state_dict_B[k]
        state_dict_A[k.replace("22", "40")] = state_dict_B[k]
        state_dict_A[k.replace("22", "47")] = state_dict_B[k]
state_dict_A.keys()
```
